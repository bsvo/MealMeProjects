```python
#get average ppg for input team
def averagePPG():
    invalidTeam = True
    while invalidTeam:
        team = input("What NBA contender do you want the average PPG for?")
        if team in conferences['East']:
            conf = 'East'
            invalidTeam = False
        elif team in conferences['West']:
            conf = 'West'
            invalidTeam = False
        else: 
            print("Invalid team, try again")
    sum = 0.0
    for player in conferences[conf][team]:
        sum += float(conferences[conf][team][player]["PPG"])
    
    return sum/(len(conferences[conf][team]))


averagePPG()
```
