```python
#imports
from lxml import etree
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import requests
from urllib.request import urlopen
#from selenium import webdriver
from bs4 import BeautifulSoup
from bs4 import Comment

```

```python
#url that links to Jerami Grant's stat page
url = "https://www.basketball-reference.com/players/g/grantje01.html"
#opens the url
html = urlopen(url)
#creates a BeautifulSoup object from it
soup = BeautifulSoup(html,"html.parser")
#soup.prettify()
#finds all tables
alltables = soup.findAll('div')
for x in range(0,len(alltables)):
    if alltables[x].findAll('tbody', limit = 15):
        print(x)
        #print(alltables[x].findAll('tbody', limit = 15))
```

```python
#Per Game

#limit indicates the max index you can take, however here we just want the first
#iterates through headers, and returns their text.
headers = [th.getText() for th in soup.findAll('tr')[0].findAll('th')]
#takes season header out
headers = headers[1:]
#finds all rows
rows = soup.findAll('tr')
#print(rows)
#Gets all cells in each row and sets them to per_game_stats 
per_game_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(0,len(rows))]
#puts the data and the column headers in a dataframe.
per_game_stats = pd.DataFrame(per_game_stats, columns = headers)
#takes out first row which is empty
per_game_stats = per_game_stats[1:11]
print("Per Game")
per_game_stats
```

```python
#totals

#takes div with all totals and sets it to narrower
narrower = alltables[0].find('div', attrs={"id":"all_totals"})
#finds the comment inside of it and sets it to comment
comment = narrower.find(string=lambda text: isinstance(text, Comment))
#narrower3 = narrower2.find('div', attrs={"id":"div_totals"})

#turns comment into new bs4 object
comment2bs4 = BeautifulSoup(comment,"html.parser")
#finds totals with data in comment2bs4
rawtotals = comment2bs4.find('div', attrs={"id":"div_totals"})
rows = rawtotals.findAll('tr')
#Gets all cells in each row and sets them to alltotalstats 
alltotalstats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(0,len(rows))]
#puts the data and the column headers in a dataframe.
alltotalstats = pd.DataFrame(alltotalstats, columns = headers)
#takes out first row which is empty
alltotalstats = alltotalstats[1:11]
print("Totals")
alltotalstats
```

```python
#Per 36 Minutes

#takes div with all totals and sets it to narrower
narrower36 = alltables[0].find('div', attrs={"id":"all_per_minute"})
#finds the comment inside of it and sets it to comment
comment36 = narrower36.find(string=lambda text: isinstance(text, Comment))
#narrower3 = narrower2.find('div', attrs={"id":"div_totals"})

#turns comment into new bs4 object
comment2bs4 = BeautifulSoup(comment36,"html.parser")
#finds per36 with data in comment2bs4
rawper36 = comment2bs4.find('div', attrs={"id":"div_per_minute"})

#limit indicates the max index you can take, however here we just want the first
#iterates through headers, and returns their text.
headers = [th.getText() for th in rawper36.findAll('tr')[0].findAll('th')]
#takes season header out
headers = headers[1:]
print(len(headers))
#makes list of rows
rows = rawper36.findAll('tr')
#Gets all cells in each row and sets them to allper36stats 
allper36stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(0,len(rows))]
#puts the data and the column headers in a dataframe.
allper36stats = pd.DataFrame(allper36stats, columns = headers)
#takes out first row which is empty
allper36stats = allper36stats[1:11]
print("Per 36 Minutes")
allper36stats
```

```python
#Per 100 Possessions

#takes div with all totals and sets it to narrower
narrower = alltables[0].find('div', attrs={"id":"all_per_poss"})
#finds the comment inside of it and sets it to comment
comment = narrower.find(string=lambda text: isinstance(text, Comment))
#narrower3 = narrower2.find('div', attrs={"id":"div_totals"})

#turns comment into new bs4 object
comment2bs4 = BeautifulSoup(comment,"html.parser")
#finds per100 with data in comment2bs4
rawper100 = comment2bs4.find('div', attrs={"id":"div_per_poss"})

#limit indicates the max index you can take, however here we just want the first
#iterates through headers, and returns their text.
headers = [th.getText() for th in rawper100.findAll('tr')[0].findAll('th')]
#takes season header out
headers = headers[1:]

#makes list of rows
rows = rawper100.findAll('tr')
#Gets all cells in each row and sets them to allper36stats 
allper100stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(0,len(rows))]
#puts the data and the column headers in a dataframe.
allper100stats = pd.DataFrame(allper100stats, columns = headers)
#takes out first row which is empty
allper100stats = allper100stats[1:11]
print("Per 100 Possessions")
allper100stats
```

```python
#Advanced

#takes div with all totals and sets it to narrower
narrower = alltables[0].find('div', attrs={"id":"all_advanced"})
#finds the comment inside of it and sets it to comment
comment = narrower.find(string=lambda text: isinstance(text, Comment))
#narrower3 = narrower2.find('div', attrs={"id":"div_totals"})

#turns comment into new bs4 object
comment2bs4 = BeautifulSoup(comment,"html.parser")
#finds per100 with data in comment2bs4
rawadvanced = comment2bs4.find('div', attrs={"id":"div_advanced"})

#limit indicates the max index you can take, however here we just want the first
#iterates through headers, and returns their text.
headers = [th.getText() for th in rawadvanced.findAll('tr', limit=4)[0].findAll('th')]
#takes season header out
headers = headers[1:]

#makes list of rows
rows = rawadvanced.findAll('tr')
#Gets all cells in each row and sets them to advancedstats 
advancedstats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(0,len(rows))]
#puts the data and the column headers in a dataframe.
advancedstats = pd.DataFrame(advancedstats, columns = headers)
#takes out first row which is empty
advancedstats = advancedstats[1:11]
print("Advanced")
advancedstats
```
