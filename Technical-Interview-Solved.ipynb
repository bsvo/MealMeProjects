```python
given_dict = {
        "menu_items": [
            {
                "name": "cake",
                "id": "001",
                "price": 1200,
                "customizations": ["523", "524"],
                "description": "Yummy Cake"
            }
        ],
        "customizations": [
            {
                "name": "flavour",
                "id": "523",
                "options": ["723", "724"],
                "description": "Choose one"
            },
            {
                "name": "size",
                "id": "524",
                "options": ["725", "726"],
                "description": "Choose one"
            }

        ],
        "options": [
            {
                "name": "chocolate",
                "id": "723",
                "price": 100,
                "description": "Milk Chocolate"
            },
            {
                "name": "vanilla",
                "id": "724",
                "price": 100,
                "description": "Milk Chocolate"
            },
            {
                "name": "small",
                "id": "725",
                "price": 25,
                "description": "Milk Chocolate"
            },
            {
                "name": "large",
                "id": "726",
                "price": 50,
                "description": "Milk Chocolate"
            }

        ]
    }
```

```python
desired_dict = {}
desired_dict["menu_items"] = {} 
menu_items  = given_dict["menu_items"]
all_customizations = given_dict["customizations"]
all_options = given_dict["options"]
#print(menu_items[0])
for item in menu_items:
    key = item["id"]
    desired_dict["menu_items"][key] = item
    Customizations = item["customizations"]
    desired_dict["menu_items"][key].pop("description", None)
    desired_dict["menu_items"][key].pop("customizations", None)
    desired_dict["menu_items"][key]["customizations"] = []
    count = -1
    for chosen_custom in all_customizations:
        if chosen_custom["id"] in Customizations:
            count+=1
            Chosen_item_customization = chosen_custom
            options = Chosen_item_customization["options"]
            Chosen_item_customization.pop("options",None)
            Chosen_item_customization.pop("description",None)
            desired_dict["menu_items"][key]["customizations"].append(Chosen_item_customization)
            desired_dict["menu_items"][key]["customizations"][count]["options"] = []
            for option in all_options:
                if option["id"] in options:
                    desired_dict["menu_items"][key]["customizations"][count]["options"].append(option)
desired_dict
```
